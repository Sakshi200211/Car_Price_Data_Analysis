```python
plt.figure(figsize=(12,5))
plt.scatter(car["Selling Price"],car["Car Condition"], marker="*" )
plt.xlabel("Selling Price")
plt.ylabel("Car Condition")
plt.title("Car Price analysis based on condition")
plt.colorbar()
plt.grid()
plt.show()
```

```python
model_fuel_type=car.groupby("Fuel Type")["Selling Price"].sum().reset_index()
```

```python
model_fuel_type
```

```python

plt.figure(figsize=(12,5))
sns.barplot(model_fuel_type,x="Fuel Type",y="Selling Price" ,palette="rocket")
plt.title("Fuel type by Model")
# plt.xticks(range(0,700,215))
plt.show()
```

```python
plt.figure(figsize=(12,5))
car.groupby("Model")["Car Condition"].value_counts().plot.barh(xlabel="Car Condition")
plt.grid(axis="x")
plt.yticks(range(1,1500,100))
plt.xticks(range(0,15,1))
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.boxplot(car["Selling Price" ],vert=False ,meanline=True ,notch=True)
plt.xlabel("Selling Price")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(car, x='Fuel Type' ,hue='Owner',palette="mako")
plt.title("Count of models by Fuel type and owner ")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(car, x="Selling Price"  ,y="Year" ,hue="Car Condition" ,palette="rocket")

plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(car,x="Selling Price",fill=True,hue="Fuel Type" ,palette="mako")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(car,x="Fuel Type" ,hue="Owner")
plt.show()
```

```python
sns.jointplot(car,x="Selling Price" ,y="Year" ,height=4,kind="hex")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.violinplot(car,x="Kilometers Driven" ,y="Owner" ,palette="rocket",fill=False )
plt.xticks(range(0,1000000,100000))
plt.title("Kilometers Driven by Owner analysis")
plt.grid()
plt.show()
```
